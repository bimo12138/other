```python
import time

class MessageInfo(object):
    def __init__(self, time_mode="%Y-%m-%d %H:%M:%S"):
        self.time_mode = time_mode
    
    def get_current_time(self):
        current_time = time.strftime(self.time_mode, time.localtime()) 
        return current_time
    
    def info(self, message):
        print("[{}] INFO: {}".format(self.get_current_time(), message))
    def success(self, message):
        print("[{}] SUCCESS: {}".format(self.get_current_time(), message))
    def warning(self, message):
        print("[{}] WARNING: {}".format(self.get_current_time(), message))
    def fail(self, message):
        print("[{}] FAIL: {}".format(self.get_current_time(), message))
        
show_message = MessageInfo()

if __name__ == "__main__":
    show_message.info("这是一个信息Message！")
    show_message.success("这是一个成功信息！")
    show_message.warning("警告信息！")
    show_message.fail("错误信息")
```

```python
# Django 新封装了 JsonResponse 类 导入方式为 from django.http import JsonResponse 封装于 HttpResponse
# 原生 return HttpResponse(json.dumps(message), content_type="application/json")
# isinstance 判断其中两个是否为一个 type

import json

class result_code(object):
    success = 200
    no_found = 404
    server_error = 500
    need_auth = 401
    server_deny = 403
    server_upload = 503
    
def result_message(http_code=result_code.success, message=""):
    message = {
        "code": http_code,
        "message": message
    }
    return json.dumps(message)
    
```
